```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.001)
```

```python
def fa(c1,x):
    return c1*np.exp(-x)+x*x-2.0*x+2.0;

def get_c1(y, x):
    return (y - x*x + 2*x -2)*np.exp(x)
```

```python
lst1 = []
with open('output.txt', 'r') as file:
    lst1 = file.read().strip().split(' ')
lst = [float(i) for i in lst1]
len(lst)
```

```python
c1 = get_c1(2, 0)
y = [fa(c1, i) for i in x]
plt.figure(figsize=(10,15))
plt.plot(x, y)
plt.plot(x, lst)
```
